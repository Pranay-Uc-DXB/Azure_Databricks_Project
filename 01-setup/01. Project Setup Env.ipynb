## Create External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS DATABRICKS_PROJECT_EXT_DL1_Formula1
URL 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL databricks_proj_sc)
COMMENT 'External location for formula1 container'
```

```sql
%fs ls 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/'
```

## Creating Catologs & Schemas

```sql
show catalogs
```

```sql
create catalog if not exists formula1
  managed location 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net'
  comment 'Main catalog for F1 Project'
  
```

```sql
CREATE SCHEMA  IF NOT EXISTS  formula1.landing;
CREATE SCHEMA  IF NOT EXISTS  formula1.Bronze
  managed location 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/Bronze';
CREATE SCHEMA  IF NOT EXISTS  formula1.Silver
  managed location 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/Silver';
CREATE SCHEMA  IF NOT EXISTS  formula1.Gold
  managed location 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/Gold';    
```

```sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@databricksprojectextdl1.dfs.core.windows.net/landing'
```
