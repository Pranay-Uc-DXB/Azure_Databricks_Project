## Driver Standings

The aim of this analytics notebook is to recreate drivers standings table just like how it is shown in the Formula 1 website below:
https://www.formula1.com/en/results/2026/drivers

To recreate the above view, following columns and tables were used:

Tables:

1 Fact table (Results)

2 Drivers table



Features:

Season,
Driver id,
driver name,
nationality,
race starts,
total points,
number of wins,
number of podiums,
standing position

```python
%sql
select * from formula1.gold.dim_results
```

```python
%sql
CREATE OR REPLACE VIEW formula1.gold.vw_driver_standings AS

select
    r.season,
    r.driver_Id,
    d.driver_name,
    d.nationality,
    SUM(r.points) AS total_points,
    COUNT(*) AS race_starts,
    count_if(r.is_win) AS number_of_wins ,
    count_if(r.is_podium) AS number_of_podiums,
    RANK() OVER (PARTITION BY r.season ORDER BY SUM(r.points) DESC) AS Position
    --r.has_points,
    --r.grid_position,
from formula1.gold.dim_results r
LEFT JOIN formula1.gold.dim_drivers d
ON r.driver_Id = d.driver_Id
GROUP BY r.season,
    r.driver_Id,
    d.driver_name,
    d.nationality
```

```python
%sql
SELECT * FROM formula1.gold.vw_driver_standings
```
