```sql
WITH Greatness_metrics AS
(SELECT 
    driver_name,
    sum(race_starts) AS Race_Starts,
    sum(total_points) AS Total_Points,
    sum(number_of_wins) AS Wins,
    sum(number_of_podiums) AS Podiums,
    sum(CASE WHEN Position=1 THEN 1 ELSE 0 END) AS championships_won
FROM formula1.gold.vw_driver_standings
GROUP BY 
    driver_name
HAVING championships_won>=1 
ORDER BY
    Total_Points DESC   
)

-- 
SELECT *,
(championships_won*100)+(wins*10)+(podiums*3) AS Greatness_score
FROM Greatness_metrics
ORDER BY
    Greatness_score DESC
```
