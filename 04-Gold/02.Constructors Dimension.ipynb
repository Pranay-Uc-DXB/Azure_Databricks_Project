## Dim Table: Races

1 Reading silver tables

2 joining circuits and races table

3 selecting relevant columns

4 Writting the final table into Gold Layer

```python
%run "../00-Common/01. Env Config"
```

```python
#Defing target table for gold layer storage
target_table=f"{catalog_name}.{gold_schema}.dim_constructors"
```

```python
# reading relevant tables form silver tables for join

constructors_df=spark.table(f"{catalog_name}.{silver_schema}.constructors")
ref_nationality_region_df=spark.table(f"{catalog_name}.{gold_schema}.ref_nationality_region")
```

```python
# joining the constructors and nationality table


dim_constructors=(
            constructors_df
            .join(
                ref_nationality_region_df,
                constructors_df.nationality == ref_nationality_region_df.nationality,
                "left")
            .select(
                constructors_df.constructor_id,
                constructors_df.constructor_name,
                constructors_df.nationality,
                ref_nationality_region_df.region.alias("nationality_region")
            )
)
```

```python
#Writting constructors data into Gold layer

(
    dim_constructors.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
)
```
